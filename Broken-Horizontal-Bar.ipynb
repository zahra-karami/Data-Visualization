```python
# create data 
jobs = {'Senior Manager':[143,192],        
        'Group Product Manager':[157,172],
        'Marketing Director' :[165,179],
        'Staff User Experience Director':[167,192],
        'Directors': [172,231],
        'Senior Partner Technology Manager': [180,195],
        'Human Resource Director': [183,199],
        'Engineering Director':[184,198],
        'Product Management Director':[203,218],
        'Lead Software Engineer Contractor':[221,239]}
```

#### The Purpose of the figure is introducing the 10 top jobs at google and the range of their salary.
Base on the the figure, It can be seen that the job with two salaries (**Senior Manager and Directors**) have the widest range of salary. while **the Lead Software Engineer Contractors** have the highest salary in google.

```python

colors = ['#fa695f','#006699','#006699','#006699','#fa695f','#006699','#006699','#006699','#006699','#006699']
fig, ax = plt.subplots(figsize=(12, 9))
ax.set_ylim(-1, 10)
ax.set_xlim(130, 250)
ax.grid(False)

for i, (key , value) in enumerate(jobs.items()):
    #set bars
    rng_start , rng_end = value[0] , value[1]-value[0]
    y_start, y_end = i-0.25 , 0.5
    ax.broken_barh([(rng_start, rng_end)], (y_start, y_end), facecolors= colors[i] )
    
    #show average of the salaries on each bar
    avg = np.mean(value)
    avg_start , avg_end = avg-.1 , .3
    ax.broken_barh([(avg_start, avg_end)], (y_start, y_end), facecolors='white' ,edgecolor='white')
    
    #add range of salary as label on both sides of bar
    ax.text(x = rng_start - 7 , y = i-0.1 , s= "${0}K".format(value[0]), fontweight='bold')
    ax.text(x = rng_start + rng_end + 1 , y = i-0.1 , s= "${0}K".format(value[1]), fontweight='bold')

# show the job titles on y axis
ax.set_yticks(list(range(len(jobs))))
ax.set_yticklabels(list(jobs.keys()), fontweight='bold', fontsize  = 12)

custom_lines = [Line2D([0], [0], color='#006699', lw=4), Line2D([0], [0], color='#fa695f', lw=4)]
ax.legend(custom_lines, ['One Salary', 'Two Salaries'],loc='upper left',)
plt.show()
```
